```python
%%manim -qh test2
class test2(Scene):
    def construct(self):
        ax=Axes()
        curve=ax.plot(lambda x: 0.01*(x**5-4*x**4+x**3-x**2),color=RED)
        x1=ax.plot(np.sin)
        area=ax.get_area(curve,x_range=(-3,0),color=WHITE)
        area1=ax.get_area(x1,x_range=(0,PI+1),color=GREEN)
        self.play(Create(ax,run_time=1),Create(curve,run_time=2),Create(x1,run_time=1))
        self.play(FadeIn(area),FadeIn(area1),run_time=2)
       
        self.wait(0.5)
        #self.add(ax,curve,area,x1,area1)
        plane=NumberPlane()
        self.play(FadeIn(plane))
        A = np.array([[-10, 1], [0, 3/4]])
        B= np.linalg.inv(A)
        x=np.pi/3
        C=np.array([[np.cos(x),-np.sin(x)],[np.sin(x),np.cos(x)]])
        g1=VGroup(curve,ax,x1,area,area1,plane)
        self.play(ApplyMatrix(A,g1))
        self.wait(0.5)
        self.play(ApplyMatrix(B,g1))
        self.play(ApplyMatrix(C,g1))
        self.play(ApplyMatrix(C,g1))
```

```python
%%manim -qh Grouping
class Grouping(Scene):
    def construct(self):
        
        plane=NumberPlane()
        self.play(Create(plane))
        d1=Dot(color=RED)
        d2=Dot(color=GREEN).next_to(d1,UR*2)
        self.play(FadeIn(d1),FadeIn(d2))
        s1=Square(color=ORANGE).shift(2*UP+4*RIGHT)
        self.play(Create(s1))
        c1=Circle(color=WHITE).move_to([-3,-2,0])
        c2=c1.copy().set_color(YELLOW).set_radius(0.5).align_to(s1,UP) 
        self.play(Create(c1),Create(c2))
        circles=VGroup(*[Circle() for _ in range(5)])
        circles.arrange(RIGHT)
        self.play(Create(circles,run_time=2))
        
```

```python
%%manim -qm Lagging
class Lagging(Scene):
    def construct(self):
        squares=VGroup(*[Square(color=RED,fill_opacity=0.5) for j in range(8) ])
        squares.arrange_in_grid(2,4).scale(0.8)
        self.play(AnimationGroup(*[FadeIn(s) for s in squares],lag_ratio=0.3))
        self.play(FadeOut(squares))

        def timevariable(mobject,t):
            radius = 4*t
            angle= 2*t*PI
            mobject.set_opacity(t)
            mobject.move_to(radius*(np.cos(angle)*RIGHT+np.sin(angle)*UP))
        s1=Square(color=WHITE).shift(RIGHT)
        s2=Square(color=RED).shift(LEFT)
        s3=Square(color=BLUE).shift(UP)
        s4=Square(color=GREEN).shift(DOWN)
        s=VGroup(s1,s2,s3,s4)
        self.play(Transform(squares,s))
        self.play(UpdateFromAlphaFunc(s1,timevariable,run_time=3,rate_func=lambda t: t**2))
        self.play(UpdateFromAlphaFunc(s2,timevariable,run_time=3,rate_func=lambda t: np.sin(t*PI)))
        self.play(UpdateFromAlphaFunc(s3,timevariable,run_time=3,rate_func=there_and_back))
        self.play(UpdateFromAlphaFunc(s4,timevariable,run_time=3,rate_func=lambda t: 1-abs(1-2*t)))
        
```
